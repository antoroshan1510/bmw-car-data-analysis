# Create countplot

```python
sns.countplot(x="model", data=df)
plt.xticks(rotation=(70))
plt.show()
```

# Create Violinplot

```python
sns.violinplot(x="transmission",y="year",data=df)
plt.show()
```

# Create pie chart

```python
labels = ['price', 'mileage', 'tax', 'mpg']
sizes = [25, 35, 20, 20]
colors = ['red', 'yellow', 'pink', 'brown']
plt.pie(sizes, labels=labels, colors=colors,
        autopct='%1.1f%%', shadow=True, startangle=140)

plt.title('BMW Car Data Analysis')
plt.axis('equal')
plt.show()
```

# Create Scatter Plot

```python
sns.scatterplot(x = "price",y = "engineSize",data = df,hue = "year")
plt.show()
```

# Create bar plot

```python
categories = ['model', 'year', 'transmission', 'fuelType']
values = [25, 40, 30, 10]
plt.bar(categories, values, color='darkcyan')
plt.title('BMW Car Data Analysis')
plt.xticks(rotation = 25)

# Show the plot
plt.show()
```

# Create Bar Plot

```python
sns.boxplot(x="transmission",y="year",data=df)
plt.title("BMW Car Data Analysis")
plt.xlabel("transmission")
plt.ylabel("year")
plt.show()
```

# Create Pair Plot

```python
sns.pairplot(df,hue="model")
plt.show()
```

```python
corr = df.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap="coolwarm")

plt.title("Correlation Heatmap")
plt.show()
```
